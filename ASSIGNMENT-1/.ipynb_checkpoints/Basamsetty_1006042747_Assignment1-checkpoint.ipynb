# Question 2C: Bootstrapping the data

```python
group= 1000 #number of times to re-sample

female_salary_bootstrap = []
male_salary_bootstrap =[]

#looping 1000 times
for i in range(group):
    #randomly selecting the samples and appending them to the list
    female_salary_avg= np.mean(female_salary_array[np.random.randint( len(female_salary_array),size= len(female_salary_array))])
    male_salary_avg= np.mean(male_salary_array[np.random.randint(len(male_salary_array),size= len(male_salary_array))])
    female_salary_bootstrap.append(female_salary_avg)
    male_salary_bootstrap.append(male_salary_avg)
```

```python
#plotting the distributions of bootstrapped data

plt.hist(female_salary_bootstrap, bins = 20,label = 'female', alpha = 0.5)
plt.hist(male_salary_bootstrap, bins = 20,label = 'male', alpha = 0.5)
plt.legend()
plt.xlabel('Salary')
plt.ylabel('frequency')
plt.title('bootstrapped distributions of Salary')
```

```python
#the below 2 blocks of code can be used to calculate and plot the difference of distributions between males and female salaries
n_salary_male= len(male_salary_bootstrap)
n_salary_female=len(female_salary_bootstrap)

mu_salary_male = np.mean(male_salary_bootstrap)
mu_salary_female = np.mean(female_salary_bootstrap)
mu_salary_diff = mu_salary_male - mu_salary_female

var_salary_male = np.var(male_salary_bootstrap)
var_salary_female = np.var(female_salary_bootstrap)
var_salary_diff = var_salary_male/n_salary_male + var_salary_female/n_salary_female
```

```python
mu = mu_salary_diff
sigma = math.sqrt(var_salary_diff)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.xlabel('difference in means')
plt.title('distribution of the difference in means of salary')
plt.show()
```

# Performing T-test on bootstrapped Data

```python
#finally by using the inbuilt function from stats package we perform the t-test on the bootstrapped data.
ttest, pval= stats.ttest_ind(female_salary_bootstrap, male_salary_bootstrap)

if pval < 0.05:
    print(pval,"Reject null hypothesis")
else:
    print(pval,"Fail to reject null hypothesis")
```

- T-test without bootstrapping the data cannot be performed as the data is not normally distributed.
- After bootstrapping the data, it looks normally distributed from the graph and hence we can perform the t-test on this data.
# Findings from the T-test
- We can observe that the P-Value(0.0) obtained from T-test is less than the threshold(0.05), so we reject null hypothesis and consider alternate hypothesis which says that the salaries of males and females are statistically different.THE GENDER DOES INFLUENCE THE SALARY OF INDIVIDUAL.

# Question 3C: Bootstrapping the data

```python
# make the result reproducable
random.seed(1000)
#ccreating a function for bootstrapping
def myBoot(sample):
    resample = random.choices(sample, k=len(sample))
    resample_mean = statistics.mean(resample)
    return resample_mean
```

```python
bachelors_salary_means = []
doctorate_salary_means = []
masters_salary_means = []

# re-sample 1000 times abd append the salaries to arrays
for i in range(1000):
    x = myBoot(df_bachelors_sal)
    bachelors_salary_means.append(x)
    
    x = myBoot(df_doctoral_sal)
    doctorate_salary_means.append(x)
    
    x = myBoot(df_masters_sal)
    masters_salary_means.append(x)
```

```python
qual_salary_data = bachelors_salary_means+doctorate_salary_means+masters_salary_means#sum of all means
```

```python
#plotting the distributions of bootstrapped data
plt.hist(bachelors_salary_means, bins = 20,label = 'bachelors', alpha = 0.5)
plt.hist(doctorate_salary_means, bins = 20,label = 'doctorate', alpha = 0.5)
plt.hist(masters_salary_means, bins = 20,label = 'masters', alpha = 0.5)
plt.legend()
plt.xlabel('Salary')
plt.ylabel('frequency')
plt.title('bootstrapped distributions of salaries grouped by qualification')
```

```python
# comparing 3 different qualifications
N = len(bachelors_salary_means)+len(doctorate_salary_means)+len(masters_salary_means)
# number of groups
k = 3

dof_between = k - 1
dof_within = N - k
```

#### Sum of Squares Within Samples (Sum of Squared Errors)

```python
SSWS = np.sum((bachelors_salary_means-np.mean(bachelors_salary_means))**2)+np.sum((doctorate_salary_means-np.mean(doctorate_salary_means))**2)+np.sum((masters_salary_means-np.mean(masters_salary_means))**2)
SSWS
```

#### Total Sum of Squares

```python
TSS = np.sum((qual_salary_data-np.mean(qual_salary_data))**2)
```

```python
TSS
```

#### Sum of Squares between Samples (Regression Sum of Squares) = TSS - SSWS

```python
SSBS = TSS-SSWS
SSBS
```

## F-Test

```python
numerator = SSBS/dof_between
denominator = SSWS/dof_within
# Calculating F fraction
F = numerator/denominator
F
```

P-value calculation

```python
p = stats.f.sf(F,dof_between,dof_within)
p
```

#### Another way to get F and p-value 

```python
f,p = stats.f_oneway(bachelors_salary_means,doctorate_salary_means,masters_salary_means)
print(f,p)
```

#### Plot Results

```python
from scipy.stats import f, norm
def plot_f_distrubiton():
    # Set figure 
    plt.figure(figsize=(12, 8))
    
    # Set degrees of freedom
    dfn, dfd = (k-1), (N-k)
    rejection_reg = f.ppf(q=.95, dfn=dfn, dfd=dfd)
    mean, var, skew, kurt = f.stats(dfn, dfd, moments='mvsk')
    
    x = np.linspace(f.ppf(0.01, dfn, dfd), 
                    f.ppf(0.99, dfn, dfd), 100)
    
    # Plot values
    plt.plot(x, f.pdf(x, dfn, dfd), alpha=0.6, 
             label=' X ~ F({}, {})'.format(dfn, dfd))
    plt.vlines(rejection_reg, 0.0, 1.0, 
               linestyles="dashdot", label="Crit. Value: {:.2f}".format(rejection_reg))
    plt.legend()
    plt.title('F-Distribution dfn:{}, dfd:{}'.format(dfn, dfd))
plot_f_distrubiton();
```

- After plotting the distributions, we can observe that the data is positive skewed, so we bootstrap the 3 groups of data and then perform ANOVA analysis.
# Findings from ANOVA
#### Conclusion: p(0.00) < 0.05 (significance level) at f value, reject null hypothethis and consider alternate hypothesis, the salaries of people having different educational qualifications are statistically different. THE EDUCATIONAL QUALIFICATION EFFECTS THE SALARY OF THE EMPLOYEE.
