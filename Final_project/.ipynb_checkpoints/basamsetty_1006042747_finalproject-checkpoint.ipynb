# Loading and Cleaning the Data

```python
df = pd.read_csv('./time_series_covid19_confirmed_global_2.csv') #https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv
df.shape #prints the shape of the dataset
```

```python
df.info()#prints information about a DataFrame including the index dtype and columns, non-null values and memory usage.
```

```python
df.describe()#Generate descriptive statistics.
```

```python
#the following code extracts the total covid cases in USA every day
# this is done by suming up the number of cases in all counties
#Since the sum obtained in cumilative SUM,daily count is extracted by subracted cases from previous day


df_us = df.loc[df['iso3']=='USA']
#Data to plot everyday total cases in USA
df_us = np.sum(df[df.columns[11:]])   #summing up the timeseries columns for entire USA
df_us_total = pd.DataFrame(df_us,columns=['counts']) 
df_us_total['counts'] = df_us_total['counts'].diff().fillna(df_us_total['counts'].iloc[0]) #calculates daily counts from cumilative SUM
df_us_total.index = pd.to_datetime(df_us_total.index,format='%m/%d/%y') #converts the index to time format which makes it easier to plot
df_us_total = df_us_total.loc[:'2020-12-10'] #limiting the end date to 10th december
#df_us_total.tail(10)
```

```python
# I also wanted to analyze COVID 19 cases in Houston
#the below code uses the same logic to extract the daily case count in Houston
#creating the final data frame for Houston and limit the end date to 10th December,2020 as mentioned in the tutorial
df_texas = df.loc[df['Province_State']=='Texas']

#let us extract the Counties in Texas with Highest number of Covid Cases as of 3rd December 2020 
df_top30 = pd.DataFrame(df_texas[['12/3/20']])
df_top30 = df_top30.set_index(df_texas['Admin2'])
df_top30 = df_top30['12/3/20'].sort_values(ascending=False)
df_top30 = df_top30.iloc[0:30]


#Extracting Houston daily Covid Cases
df_houston = df_texas[df_texas['Admin2'] == 'Houston']
df_houston_final = pd.DataFrame(df_houston[df_houston.columns[11:]].sum(),columns=["counts"])
df_houston_final['counts'] = df_houston_final['counts'].diff().fillna(df_houston_final['counts'].iloc[0])
df_houston_final[df_houston_final['counts']<0]=0
df_houston_final.index = pd.to_datetime(df_houston_final.index,format='%m/%d/%y')
df_houston_final = df_houston_final.loc[:'2020-12-10']
#df_houston_final.tail(30)
```

# Modelling

# ARIMA model
- ARIMA, short for â€˜Auto Regressive Integrated Moving Averageâ€™ is actually a class of models that â€˜explainsâ€™ a given time series based on its own past values, that is, its own lags and the lagged forecast errors, so that equation can be used to forecast future values.
- An ARIMA model is characterized by 3 terms: p, d, q where


1. p is the order of the AR term.

2. q is the order of the MA term.

3. d is the number of differencing required to make the time series stationary.

Since ARIMA model has a fixed structure and is specifically built for time series (sequential) data, I have chosen this model.

# Augmented Dickey Fuller Test ( ADF Test )
- This test is also to check stationary of time series. And even we can use to find out the number of differencing used on ARIMA model for forecasting. This test assumes null hypothesis that the series is non-stationary

```python
#augmented Dickey Fuller test on US values
result = adfuller(df_us_total.values, autolag='AIC')
print(f'ADF Statistic: {result[0]}')
print(f'n_lags: {result[1]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}') 
```

- The static value is greater than the critical value and also the p-value is greater than the threshold value(0.05), hence the time series for USA is non stationary.

```python
#plotting the mid case projections
arima_model_us = ARIMA(df_us_total,order=(4,3,3))
res_us = arima_model_us.fit()
forecast_us_1 = res_us.predict(start=len(df_us_total),end=400,n_periods=len(df_us_total))

fig = plt.figure(figsize=(11,6))
plt.plot(df_us_total.index,df_us_total.values,marker='o', label='Daily Covid Cases in USA From Jan to Dec')
plt.plot( forecast_us_1.index ,forecast_us_1.values  ,color='orange',label='Mid case Projection')
ax.xaxis.set_minor_locator(mdates.DayLocator(interval=1))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Total Number of Cases')
plt.title('Mid Covid Case projections in USA')
plt.rcParams.update({'font.size': 22})
plt.legend()
```

```python
#plotting the best case projections where the cases tend to decrease over time
arima_model_us = ARIMA(df_us_total,order=(6,0,0))
res_us = arima_model_us.fit()
forecast_us_2 = res_us.predict(start=len(df_us_total),end=400,n_periods=len(df_us_total))
fig = plt.figure(figsize=(11,6))
plt.plot(df_us_total.index,df_us_total.values, marker='o', label='Covid Cases in USA From Jan to Dec')
plt.plot( forecast_us_2.index ,forecast_us_2.values , color='green',label='Projection for best cases')
ax.xaxis.set_minor_locator(mdates.DayLocator(interval=1))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Total Number of Cases')
plt.title('Best Covid Case projections in USA')
plt.rcParams.update({'font.size': 22})
plt.legend()
```

```python
#plotting the worse case projections where there is an exponential rise in case count
arima_model_us = ARIMA(df_us_total,order=(4,3,1))
res = arima_model_us.fit()
forecast_us_3 = res.predict(start=len(df_us_total),end=400,n_periods=len(df_us_total))
fig = plt.figure(figsize=(11,6))
plt.plot(df_us_total.index,df_us_total.values, marker='o', label='Covid Cases in USA From Jan to Dec')
plt.plot( forecast_us_3.index ,forecast_us_3.values , color='red',label='Worst Case Projection in USA')
ax.xaxis.set_minor_locator(mdates.DayLocator(interval=1))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Total Number of Cases')
plt.title('Worst Covid Case projections in USA')
plt.rcParams.update({'font.size': 22})
plt.legend()
```

# ARIMA Model For Houston Covid Cases

```python
# Projection for best cases in Houston where the green line represents decreases in cases
arima_model = ARIMA(df_houston_final,order=(1,1,5))
res = arima_model.fit()
forecast_2 = res.predict(start=len(df_houston_final),end=400,n_periods=len(df_houston_final))
fig = plt.figure(figsize=(11,6))
plt.plot(df_houston_final.index,df_houston_final.values, marker='o', label='Covid Cases in Houston From Jan to Dec')
plt.plot( forecast_2.index ,forecast_2.values , color='green',label='Projection for Best cases')
ax.xaxis.set_minor_locator(mdates.DayLocator(interval=1))
ax.xaxis.set_major_locator(mdates.DayLocator(interval=5))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d'))
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.ylabel('Total Number of Cases')
plt.legend()
```

# Correlating stay home orders in Houston with Population violating the orders
