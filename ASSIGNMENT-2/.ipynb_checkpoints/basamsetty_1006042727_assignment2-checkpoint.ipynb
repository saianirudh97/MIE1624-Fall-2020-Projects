# Question 1: Cleaning the Data

## Remove Unnecessary Columns

The undesired columns in our dataset are removed using the below code. The columns with "OTHER_TEXT" suffix will be removed and since we will be using 'Q10_encoded' as the target column, all other columns with 'Q10' prefix i.e  'Q10_buckets' and 'Q10' are removed.

## Encode Multi-parts Questions

Since, these question are multiple choice questions and consists of fixed number of choices and participants may also skip the question which leads to NAN values.
The encoding of this type of questions is done using following steps:

1. Replace selected options by 1 and non-selected options by 0.
2. Sum up the selected options (except for the 'None' option) of each participant for each question and create a new feature with the sum.


```python
numofentries = df.shape[0]

def multiple_choice(col,numofparts):
  col_index = df.columns.get_loc(col+'_Part_'+str(numofparts))
  df.insert(col_index+1, col+'_sum', [0]*numofentries, True)
  for i in range(numofparts):
    df[col+'_Part_'+str(i+1)]=df[col+'_Part_'+str(i+1)].notnull().astype(int)
    if i != numofparts-2:
      df[col+'_sum'] = df[col+'_sum'] + df[col+'_Part_'+str(i+1)]
  return df[col+'_sum'] 
```

```python
Q9_sum = multiple_choice('Q9',8)
Q12_sum = multiple_choice('Q12',12)
Q13_sum = multiple_choice('Q13',12)
Q16_sum = multiple_choice('Q16',12)
Q17_sum = multiple_choice('Q17',12)
Q18_sum = multiple_choice('Q18',12)
Q20_sum = multiple_choice('Q20',12)
Q21_sum = multiple_choice('Q21',5)
Q24_sum = multiple_choice('Q24',12)
Q25_sum = multiple_choice('Q25',8)
Q26_sum = multiple_choice('Q26',7)
Q27_sum = multiple_choice('Q27',6)
Q28_sum = multiple_choice('Q28',12)
Q29_sum = multiple_choice('Q29',12)
Q30_sum = multiple_choice('Q30',12)
Q31_sum = multiple_choice('Q31',12)
Q32_sum = multiple_choice('Q32',12)
Q33_sum = multiple_choice('Q33',12)
Q34_sum = multiple_choice('Q34',12)
```

# Question 2B: Feature Selection

Feature engineering is the process of transforming raw data into features that better represents the underlying problem to the predictive models, resulting in improved accuracy on unseen or test data. This process consists of feature importance,  feature selection, feature extraction, and feature construction.

Irrelevant or partially relevant features can negatively impact model performance and that is why we need feature selection. 

## Feature Selection using Random Forest

Random forests are one the most popular machine learning algorithms, which provides a good predictive performance, low overfitting, and easy interpretability.

Random forests consists of 4 to 1200 decision trees, each of them built over a random extraction of the observations from the dataset and a random extraction of the features. Not every tree sees all the features or all the observations, and this guarantees that the trees are de-correlated and therefore less prone to over-fitting. Each tree is also a sequence of yes-no questions based on a single or combination of features. At each node, the tree divides the dataset into 2 buckets, each of them hosting observations that are more similar among themselves and different from the ones in the other bucket. Therefore, the importance of each feature is derived from how “pure” each of the buckets is.

```python
y = df['Q10_Encoded']
X = df.drop(['Q10_Encoded'],axis=1)
```

```python
scaler = StandardScaler()
df_std = scaler.fit_transform(X)
df_std_ = pd.DataFrame(df_std)
```

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble.forest import RandomForestClassifier
from sklearn import model_selection
```

```python
select = SelectFromModel(RandomForestClassifier(n_estimators = 10))
select.fit(df_std_, y)
```

```python
select_feat = df_std_.columns[(select.get_support())]
print('total features: {}'.format((df_std.shape[1])))
print('selected features: {}'.format(len(select_feat)))
```

```python
select_list=[]
for i in range(len(select_feat)):
    select_list.append(select_feat[i])

df_RF = df_std_[df_std_.columns.intersection(select_list)]
```

```python
print('Number of Features left: ', df_RF.shape[1])
```

## Further Feature Selection using Principle Component Analysis (PCA)

The number of remaining features is still high (around 120), in this case, Principle Component Analysis is used to further reduce the dimensions of the dataset.

```python
# Further reduce the features in half and obtain 60 features left
from sklearn.decomposition import PCA
pca = PCA(n_components = 60)
pca.fit(df_RF)
X_sel = pca.fit_transform(df_RF)

#convert it to dfframe
X_sel = pd.DataFrame(X_sel)
```

```python
print('Number of Features left: ', X_sel.shape[1])
```

# Question 3: Model Implementation

## Split Training and Test Set

The entire data is split into 70% training and 30% test sets.

```python
X_train, X_test, y_train, y_test = train_test_split(X_sel, y, random_state=1, test_size=0.3)
```

## Logistic Regression using 10-Fold Cross-Validation

Cross-validation is a resampling procedure used to evaluate machine learning models on a limited data.

This procedure has a parameter k that refers to the number of groups that a given data sample can be split into. 

Cross-validation is primarily used in applied machine learning to estimate the skill of a machine learning model on unseen data. That is, to use a limited sample in order to estimate how the model is expected to perform in general when used to make predictions on data not used during the training of the model. It generally results in a less biased or less optimistic estimate of the model skill.

```python
# Obtain logistic regression and cross validation scores
LR_model = LogisticRegression(penalty='l1',C=1.0,solver='liblinear') #use logistic model with L1
results = cross_val_score(LR_model, X_train, y_train, cv=10) #using 10 folds
```

```python
# Print accuracy results for each fold
for i in range(10):
    score = results[i] * 100
    print("Fold "+str(i+1)+" Accuracy: "+str(round(score,3))+"%")

# Print average accuracy and variance
score_avg = results.mean()*100
score_std = results.std()*100
score_var = pow(score_std,2)
print("Average Accuracy: "+str(round(score_avg,3))+"%") 
print("Variance of Accuracy: "+str(round(score_var,3))+"%")
```

## Treating Each Value of Hyperparameter as a New Model

In this section, 12 values of parameter C strength are implemented in 12 models to obtain their scores.

```python
c = [0.001,.009,0.01,.09,1,5,10,25,50,100,500,1000]
for i in range(12):
    model= LogisticRegression(penalty='l1',C=c[i],solver='liblinear')
    model.fit(X_train, y_train)
    accuracy = round(model.score(X_test, y_test),3)
    print("Model",i+1,"Score:",accuracy)
```

The model with highest score of 0.348 corresponds to a C value of 1.0.

# Question 4: Model Tuning

## Grid Search

In this section, the hyperparameter C and regularization method (L1 or L2) are tuned to find the optimal model. Grid search CV helps in hyper parameter tuning. We use 10 folds as specified in the question.

```python
clf = LogisticRegression(solver='liblinear')
grid_values = {'penalty': ['l1', 'l2'],'C':[0.001,.009,0.01,.09,1,5,10,25,50,100,500,1000]}
grid_clf_acc = GridSearchCV(clf,param_grid = grid_values,cv=10)
optimal = grid_clf_acc.fit(X_train, y_train)
```

```python
print('C:', optimal.best_estimator_.get_params()['C'])
print('Penalty:', optimal.best_estimator_.get_params()['penalty'])
print("Accuracy :",optimal.best_score_)
```

```python
final_model = LogisticRegression(penalty='l1',C=1.0, solver='liblinear')
final_model.fit(X_train, y_train)
```

# Question 5: Testing and Discussion

## Testing of the model

The model is run on the test set and the results are displayed as below.

```python
y_pred = final_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```

The precision, recall and f1-score are listed in the table above. The overall accuracy is 0.34 which generally aligns with that of the training set, only slightly lower. Classes with more support tend to have higher precision, recall and f1 score.
